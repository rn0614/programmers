```python
def TCD(folder,file):
    data = pd.read_csv(f'{folder}/{file}.dat', sep='|', encoding='ANSI', header=None)
    data.columns=TCD_colnames
    data.to_csv(f'{file}.csv', encoding='ANSI')
```

```python
# dat파일 csv 파일로 변경
def sttn_read(file):
    # dat 파일 읽기  head=None은 dat 파일에 컬럼명이 없는 것을 의미
    df_sttn = pd.read_csv(file, sep='|', encoding='ANSI', header=None)
    STTN_colnames=['운행일자','정산사코드','정산 지역 코드','정류장 ID','정류장 명칭',
              '정류장 ARS번호','정류장 X 좌표', '정류장 Y 좌표']# 전체 column 명 지정
    df_sttn.columns=STTN_colnames
    # 일부 column 명 사용
    df_sttn=df_sttn[['정류장 ID','정류장 명칭','정류장 X 좌표','정류장 Y 좌표']]
    return df_sttn

# 바꿀 모든 날자 입력과정
names=['STTN_20191021.dat','STTN_20191023.dat','STTN_20191024.dat','STTN_20191025.dat','STTN_20191026.dat','STTN_20191027.dat','STTN_20191028.dat','STTN_20201019.dat','STTN_20201020.dat','STTN_20201021.dat','STTN_20201022.dat','STTN_20201023.dat','STTN_20201024.dat','STTN_20201025.dat','STTN_20210419.dat','STTN_20210420.dat','STTN_20210421.dat','STTN_20210422.dat','STTN_20210423.dat','STTN_20210424.dat','STTN_20210425.dat']
#names=['DWTCD_20210425.dat'] #허루만 볼때
df_sttn=pd.DataFrame()
for i in names:
    df_sum=sttn_read(i)
    df_sttn=pd.concat([df_tcd,df_sum])
```
